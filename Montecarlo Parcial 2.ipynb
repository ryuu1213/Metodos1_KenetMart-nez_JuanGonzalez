```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d
from scipy import integrate
from tqdm import tqdm
```

```python
def CreateSphere(N,R=1):
    
    Points = np.zeros((N,3))
        
    for i in tqdm(range(N)):
        
        phi = np.random.uniform(0,2*np.pi)
        u = np.random.rand()
        r = R*u**(1/3)
        costheta = np.random.uniform(-1,1)
        theta = np.arccos(costheta)
        
        Points[i] = [r*np.cos(phi)*np.sin(theta),r*np.sin(phi)*np.sin(theta),r*np.cos(theta)]
        
    return Points
```

```python
N = 100000
Sphere = CreateSphere(N)
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(Sphere[:,0],Sphere[:,1],Sphere[:,2])
```

```python
#montecarlo  1
# Aaverage of f(x, y)
Ixx = np.average(Sphere[:,1]**2 + Sphere[:,2]**2)
Iyy = np.average(Sphere[:,0]**2 + Sphere[:,2]**2)
Izz = np.average(Sphere[:,0]**2 + Sphere[:,1]**2)
        
print(Ixx, Iyy, Izz)
```

```python
#Montecarlo 2
Ixy = -1*np.average(Sphere[:,0] * Sphere[:,1])
print(Ixy)

"""
	La estimación del valor de la rotación en los ejes de simetrías combinados, en este caso, Ixy, tiende a 0 ya que la distribución
    de masa de la esfera es uniforme radialmente. Para que Ixy no diera cero, tendría que haber una concentración de masa sobre el eje x=y
    o algo parecido.
"""
```
